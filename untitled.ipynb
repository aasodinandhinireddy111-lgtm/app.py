```python
#step3: Handling the missing data or values
#scenario-1:
#scenario-2:
#scenario-3:

import numpy as np
from sklearn.impute import SimpleImputer

imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(x[:,1:3])
x[:,1:3]=imputer.transform(x[:,1:3])
```

```python
#step4: Handling the catogorial data
#which coverts the text data into numeric data.
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

#Alexa-0,0,1

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')#endcoder is a method which converts text to binary
np.remainder='passthrough'
x=ct.fit_transform(x)
```
